```python
## DataFrame 생성
df_1 = pd.read_csv('result.csv', names = ['title','date','url','script'])
print(df_1)
df_2 = pd.read_csv('result.csv', names = ['title','date','url','script'])

date=''
prepro=''
cnt= 1
for k in range(len(tmp)):
    for i in tmp[k]:
        #제목
        #print("제목 : " + i.get_text())
        title = i.get_text()
        #URL
        url = str(i['href'])
        response = requests.get(url)
        soup = BeautifulSoup(response.text,'html.parser')
        #print("URL : " + url)
        #날짜
        if soup.find('div', class_='tt') is not None:
            date = soup.find('div', class_='tt').find('em')
            date = str(date)
        #print("date : "+ date)
        #본문
        if soup.find('div', class_='article') is not None:
            div_tag = soup.find('div', class_='article')
            prepro = str(div_tag)
            prepro = re.sub('\n','',prepro,0).strip()
            prepro = re.sub('<!--.+?-->','',prepro,0).strip()
            prepro = re.sub('<.+?>','',prepro,0).strip()
        #print(prepro)
        #print('',end='\n\n')
        new_df = pd.DataFrame({'title':title, 'date':date, 'url':url, 'script':prepro}, index=[0])
        df_1 = pd.concat([df_1, new_df])
        df_2.append(new_df)
```
